```python
import random

def random_swap(sentence, n=2):
    sentence = sentence.split () 
    length = range(len(sentence))
    swapped = []
    for _ in range(n):
        idx1, idx2 = random.sample(length, 2)
        swapped.append ([sentence[idx1], sentence[idx2]])
        sentence[idx1], sentence[idx2] = sentence[idx2], sentence[idx1] 
    return ' '.join (sentence), swapped
```

```python
def create_mdistilbert_model(mdistilbert_model):
    input_ids = tf.keras.Input(shape=(150,),dtype='int32')
    attention_masks = tf.keras.Input(shape=(150,),dtype='int32')
    
    losss ='sparse_categorical_crossentropy'

    weight_initializer = tf.keras.initializers.GlorotNormal(seed=42) 
    last_hidden_state = mdistilbert_model([input_ids, attention_masks])[0]

    cls_token = last_hidden_state[:, 0, :]
    
    output = tf.keras.layers.Dense(5, 
                                   activation='softmax',
                                   kernel_initializer=weight_initializer,  
                                   kernel_constraint=None,
                                   bias_initializer='zeros'
                                   )(cls_token)
    
    model = tf.keras.Model([input_ids, attention_masks], output)
    
    model.compile(tf.keras.optimizers.Adam(learning_rate=1.4e-6), 
                  loss=losss,
                  metrics=['accuracy'])
    
    return model
```

```python

def create_mbert_model(mbert_model):
    input_ids = tf.keras.Input(shape=(150,),dtype='int32')
    attention_masks = tf.keras.Input(shape=(150,),dtype='int32')
    
    output = mbert_model([input_ids,attention_masks])
    output = output[1]
    
    output = tf.keras.layers.Dense(16,activation='relu')(output)
    output = tf.keras.layers.Dropout(0.5)(output)

    output = tf.keras.layers.Dense(5,activation='softmax')(output)
    model = tf.keras.models.Model(inputs = [input_ids,attention_masks],outputs = output)
    model.compile(Adam(learning_rate=5e-6), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python


def create_model_xlnet(xlnet_model):
    word_inputs = tf.keras.Input(shape=(150,), name='word_inputs', dtype='int32')

    
    xlnet = TFXLNetModel.from_pretrained(xlnet_model)
    xlnet_encodings = xlnet(word_inputs)[0]

    # Collect last step from last hidden state (CLS)
    doc_encoding = tf.squeeze(xlnet_encodings[:, -1:, :], axis=1)
    
    doc_encoding = tf.keras.layers.Dropout(.1)(doc_encoding)
     
    outputs = tf.keras.layers.Dense(5, activation='softmax', name='outputs')(doc_encoding)

    model = tf.keras.Model(inputs=[word_inputs], outputs=[outputs])
    model.compile(optimizer=tf.keras.optimizers.Adam(lr=1e-5), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return model

```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix, classification_report, matthews_corrcoef
from collections import Counter
import math
ens_accuracy=[]
ens_mcc=[]
ens_prec = []
ens_reca =[]
ens_fs=[]
# ens_cr=[]
for fold in range(0,2):
    mx_vote_df = pd.DataFrame()
    mx_vote_df['target'] = test['target']
    
    
#     missing_val_count_by_column = (mx_vote_df['target'].isnull().sum())
#     print(missing_val_count_by_column)
#     mx_vote_df['target'] = mx_vote_df['target'].dropna(axis=0)
#     mx_vote_df['target'] = (mx_vote_df['target'].isnull().sum())
#     print(mx_vote_df['target'])


    mx_vote_df['bert'] = pred_bert[fold]
    mx_vote_df['xlnet'] = pred_xl[fold]
    mx_vote_df['mbert'] = pred_mbert[fold]
    mx_vote_df['distilbert'] = pred_distilbert[fold]
    mx_vote_df['mdistilbert'] = pred_mdistilbert[fold]
    
    mx_vote_df = mx_vote_df.dropna(axis=0)
    # vote_df.append(mx_vote_df)
    print('--- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---')
    print(mx_vote_df[:50])

#     print("bert = \n", mx_vote_df['bert'].value_counts())
#     print("voting mbert = \n", mx_vote_df['mbert'].value_counts())
#     print("voting distilbert = \n",mx_vote_df['distilbert'].value_counts()) 

    mx_vote_df = mx_vote_df[['target','bert','mbert', 'distilbert', 'mdistilbert','xlnet']]

    preds = []
    for index in range(len(mx_vote_df)):
        values = mx_vote_df.iloc[index].values[1:]
        sorted_x = sorted(Counter(values).items(), key=lambda kv: kv[1],reverse=True)
        preds.append(sorted_x[0][0])
    # preds.append(math.floor(sum(values)/5))
    mx_vote_df['preds'] = preds
    print(mx_vote_df[:50])


    y_test = mx_vote_df.target.values
    y_pred = mx_vote_df['preds'].values

    ens_accuracy.append(metrics.accuracy_score(y_test, y_pred))
    ens_mcc.append(matthews_corrcoef(y_test, y_pred))
    ens_prec.append( metrics.precision_score(y_test, y_pred, average='weighted'))
    ens_reca.append( metrics.recall_score(y_test, y_pred, average='weighted'))
    ens_fs.append(metrics.f1_score(y_test, y_pred, average='weighted'))

    print(f'Fold {fold} :: Ensemble Accuracy ::', metrics.accuracy_score(y_test, y_pred))
    print(f'Fold {fold} :: Ensemble Mcc Score::', matthews_corrcoef(y_test, y_pred))
    print(f'Fold {fold} :: Ensemble Precision::', metrics.precision_score(y_test, y_pred, average='weighted'))
    print(f'Fold {fold} :: Ensemble Recall::', metrics.recall_score(y_test, y_pred, average='weighted'))
    print(f'Fold {fold} :: Ensemble F_score::', metrics.f1_score(y_test, y_pred, average='weighted'))
    print(f'Fold {fold} :: Ensemble classification_report:: \n', metrics.classification_report(y_test, y_pred))
    print('--- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---')
    print('--- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---')

```
